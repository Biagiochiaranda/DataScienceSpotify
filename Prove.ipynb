## Confronto canzoni in comune

```
def common_songs_number(list1, list2):
    common_count = 0
    if len(list1) < len(list2):
        shorter_list = list1
        longer_list = list2
    else:
        shorter_list = list2
        longer_list = list1
    
    for element in shorter_list:
        if element in longer_list:
            common_count += 1
    
    return common_count
```
