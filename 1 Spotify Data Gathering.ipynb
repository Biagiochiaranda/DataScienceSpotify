- Ottenimento delle categorie con chiamata all'endpoint **Get Several Browse Categories** `/browse/categories`:

```
result_categories = spotify.categories(country='IT', locale='it_IT', limit=50, offset=0)
r = result_categories['categories']
print(f'totale categorie: {r["total"]}')
```

```
ids = []
names = []
for item in r['items']:
    ids.append(item['id'])
    names.append(item['name'])
print(f'id: {ids}')
print(f'names: {names}')
```

- Raccolta delle playlist di ogni categoria come lista di dizionari:

```
categories_playlists = []
for index, row in df.iterrows():
    playlists = []
    try:
        r = spotify.category_playlists(category_id=row['ID'],country='IT',limit=50)
        for p in r['playlists']['items']:
            try:
                playlists.append(p['id'])
            except:
                pass
            while r['playlists']['next']:
                r = spotify.next(r['playlists'])
                for p in r['playlists']['items']:
                    try:
                        playlists.append(p['id'])
                    except:
                        pass
    except:
        pass
    categories_playlists.append({'NAME_CATEGORIES': row['Nome'], 'ID_CATEGORIES': row['ID'], 'ID_PLAYLISTS': playlists, 'N_PLAYLISTS': len(playlists)}) 
```

```
l = playlist_df['ID_PLAYLISTS']
id_list = []
for lista in l:
    #print(type(lista))
    row = lista.replace('[', '').replace(']', '').replace('"', '').replace("'", '').replace(' ', '')
    row = row.split(',')
    id_list.append(row)
#id_list[1]
```

```
playlists_tracks = []
for index, row in playlist_df.iterrows():
    for p_id in id_list[index]:
        tracks = []
        #print(p_id)
        try:
            playlist = spotify.playlist(p_id)
            name = playlist['name']
            total = playlist['tracks']['total']
            r = spotify.playlist_items(p_id)
            #print(r)
            for t in r['items']:
                try:
                    tracks.append(t['track']['id'])
                except:
                    pass
                while r['next']:
                    r = spotify.next(r)
                    for t in r['items']:
                        try:
                            tracks.append(t['track']['id'])
                        except:
                            pass
            playlists_tracks.append({'ID_PLAYLIST': p_id, 'PLAYLIST_NAME': name, 'ID_TRACKS': tracks, 'N_TRACKS': len(tracks), 'TOTAL': total}) 
        except:
            pass
# N.B.: ESECUZIONE DURATA BEN 29 MINUTI!!!! (Me Cojoni)
```

```
tracks_df = pd.DataFrame(playlists_tracks)
tracks_df
```

### Grafo

```
import networkx as nx

# Definizione delle due liste
list1 = tracks_df['ID_PLAYLIST'] #lista di id di playlist
list2 = l2 #lista di id di canzoni
```

```
len(list2)
```

```
# Creazione del grafo
G = nx.Graph()

# Aggiunta dei nodi dal primo elenco
G.add_nodes_from(list1)

for i in range(len(list2)):
    node1 = list2[i]
    for j in range(i+1, len(list2)):
        node2 = list2[j]
        if node1 != node2:
            common = set(node1) & set(node2)
            if common:
                G.add_edge(node1, node2, weight=len(common))

# Visualizzazione del grafo
nx.draw(G, with_labels=True)
```
